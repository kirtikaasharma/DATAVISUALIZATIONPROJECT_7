```python
import pandas as pd
import scipy 
from scipy import stats
df = pd.read_csv("cutlets.csv");
df

```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df.columns = "Unit_A", "Unit_B"
df
```

```python
# Normality Test
# Ho: there is no difference in diameter of the cutlets
# Ha= there is difference in diameter
stats.shapiro(df.Unit_A) # Shapiro Test

# since p-value >.05 so Null hypothesis will be considered Hence normal
```

```python
stats.shapiro(df.Unit_B)
```

```python
# Variance test
#HO: Variance are equal  #Ha:variances are not equal
scipy.stats.levene(df.Unit_A, df.Unit_B)
```

```python
scipy.stats.ttest_ind(df.Unit_A,df.Unit_B)
```
